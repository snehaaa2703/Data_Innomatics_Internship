1. Problem Statement:


   Given a username and password, check whether login is successful.


```python
def login():
  # declaring the valid username and password
  valid_username="admin"
  valid_password="1234"

#  Asking for user input
  user=input("Enter username: ")
  password=input("Enter password: ")

#  validating empty inputs and
# Asking the users to input both username and password if anyone of them is empty then it will return false
  if user=="" or password=="":
     print("Enter both username and password: ")
     return False

# Checking if both username and password are valid if it is valid then only it will print the successful login message or else will print invalid credentials
  elif user==valid_username and password==valid_password:
     print("Login Successful")
     return True

  else :
    print("Invalid Credentials")
    return False

# Calling the function to execute
login()
```

2. Problem Statement:


   Analyze student results. marks = [45, 78, 90, 33, 60]


```python
# list of marks
marks=[45,78,90,33,60]
def resultAnalyzer(marks):
  passed= 0
  failed= 0

# checking the marks and segregating it based on condition
  for i in marks:
    if i>=50:
      passed+=1
    else:
      failed+=1

  return passed, failed

# making function call here and storing it in variable
passed,failed=resultAnalyzer(marks)

# printing the stored result here
print("Total no. of students passed:",passed)
print("Total no. of students failed:",failed)
```

3. Problem Statement:

   Clean and standardize user names. names = [" Alice ", "bob", " CHARLIE "]

```python
names = [" Alice ", "bob", " CHARLIE "]

def clean(names):

  # definig empty list to store the cleaned names
  clean_list=[]

  for i in names:

 # strip() will remove the extra spaces and lower() will convert to lowercase
    cleaned_name=i.strip().lower()

    # cleaned name will be added to the emptylist by append()
    clean_list.append(cleaned_name)

  print("cleaned names: ",clean_list)

# making the function call
clean(names)
```

4. Problem Statement:


Analyze message sizes.

messages = ["Hi", "Welcome to the platform", "OK"]


```python
messages = ["Hi", "Welcome to the platform", "OK"]
def lenAnalyzer(messages):

  for i in messages:
    # len() will count the length of msg
    length=len(i)
    print("length of",i,"is:",length)

    if length>10:
      print("messages longer the 10 characters")
    else:
      print("Everything is fine")
    print("--------------------")

lenAnalyzer(messages)
```

5. Problem Statement:


   Detect error messages from system logs.

   logs = ["INFO", "ERROR", "WARNING", "ERROR"]


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
def errorCount(logs):
  # considering the error count as 0
  count=0

  # iterating through the logs
  for i in logs:
    if i=="ERROR":
      count+=1

  print("Total no. of Error entries are: ",count)

errorCount(logs)
```
