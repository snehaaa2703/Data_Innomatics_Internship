1. Problem Statement

   Create a function to validate a mobile recharge request.


```python
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Amount must be at least ₹50"
    if plan not in valid_plans:
        return "Invalid plan selected"
    if amount != plan:
        return "Amount must match selected plan"

    return "success"


while True:
    amount = int(input("Enter recharge amount: "))
    plan = int(input("Enter plan: "))

    result = validate_recharge(amount, plan)

    if result == "success":
        print("Recharge successful")
        break
    else:
        print("Error:", result)
        print("Please try again.\n")

```

2. Problem Statement

    Inventory Reorder Alert System

```python
def check_stock(products):
    for item, qty in products.items():
        if qty < 15:
            print(item, "→ Reorder Alert")
        else:
            print(item, "→ Stock OK")


# Take user input
n = int(input("Enter number of products: "))
inventory = {}

for _ in range(n):
    name = input("Enter product name: ")
    qty = int(input(f"Enter stock quantity for {name}: "))
    inventory[name] = qty

# Check stock
check_stock(inventory)
```

3. Problem :

   Student Result Processing System


```python
def check_pass(marks):
    total = 0
    for mark in marks:
        total += mark  # sum all marks
    average = total / len(marks)  # calculate average

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example usage
n = int(input("Enter number of students: "))
marks = []

for _ in range(n):
    m = float(input("Enter mark: "))
    marks.append(m)3

result = check_pass(marks)
print(result)
```

4. Problem :

   Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak=False):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak:
        fare *= 1.25  # 25% extra

    return fare


while True:
    try:
        distance = float(input("Enter distance (km): "))
        peak_input = input("Is it peak hour? (yes/no): ").strip().lower()
        peak = True if peak_input == 'yes' else False

        total_fare = calculate_fare(distance, peak)
        print(f"Total Fare: ₹{total_fare:.2f}")
        break  # exit loop after successful calculation

    except ValueError:
        print("Invalid input. Please enter numbers only. Try again.\n")
```

5. Problem :

   Employee Attendance Eligibility Checker


```python
def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day.upper() == "P":
            present_days += 1

    # Calculate percentage
    percentage = (present_days / total_days) * 100

    # Return eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example usage
n = int(input("Enter number of days: "))
attendance = []

for i in range(n):
    status = input(f"Day {i+1} (P/A): ").strip().upper()
    attendance.append(status)

result = check_attendance(attendance)
print(result)
```

6. Problem :

   Password Strength Checker


```python
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    # Loop over string to check conditions
    for ch in password:
        if ch.isdigit():          # check for digit
            has_digit = True
        if ch in special_chars:   # check for special character
            has_special = True

    # Check minimum length
    if len(password) < 8:
        return "Weak: Less than 8 characters"
    if not has_digit:
        return "Weak: No digit"
    if not has_special:
        return "Weak: No special character (@#$)"

    return "Strong"


# Retry until valid password
while True:
    pwd = input("Enter password: ")
    result = check_password(pwd)
    print(result)

    if result == "Strong":
        break
    else:
        print("Try again.\n")
```
