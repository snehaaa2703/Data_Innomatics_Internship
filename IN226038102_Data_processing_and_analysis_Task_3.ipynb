1. Problem Statement :

    Employee Performance Bonus Eligibility


```python
employees={}

# no. of employees
n=int(input("Enter no. of Employees: "))

# taking user input
for i in range(n):
  name=input("Enter employee name: ")
  score=int(input("Enter performance score: "))
  employees[name]=score

# step 1:for Find Highest score
  highest_score=max(employees.values())

# step 2:Finding employees with highest score
performers=[name for name, score in employees.items() if score == highest_score]

# step 3: For displaying result
print("Top Performers Eligible for bonus: ",", ".join(performers),f"(score: {highest_score})")
```

2. Problem Statement :

   Search Query Keyword Analysis


```python
import string
# Taking user input
text=input("Enter your query: ")

# step 1: for converting into lowercase using lower()
text=text.lower()

# step 2: to remove punctuation
text=text.translate(str.maketrans('','',string.punctuation))

# step 3: for splitting the words
words=text.split()

# step 4: count frequency
freq={}
for w in words:
  freq[w] =freq.get(w,0)+1

# step 5: displaying only keywords searched more than once
result={w:count for w, count in freq.items() if count>1}

print(result)
```

3. Problem Statement:
   
   Sensor Data Validation


```python
# taking user input for total no. of readings
n=int(input("Enter no. of sensor reading: "))

# to store
reading=[]

# taking readings from user
for i in range(n):
  value=int(input(f"Enter reading for hour {i}: "))
  reading.append(value)

# identify valid(even) readings
valid_reading=[(index,value) for index, value in enumerate(reading) if value%2==0]

# display result
print("Valid sensor readings (hour,value): ")
print(valid_reading)
```

4. Problem Statement :

   Email Domain Usage Analysis

```python
# list to store mails
emails= []

n=int(input("Enter no. of emails: "))

# taking emails from user
for i in range(n):
  e=input("Enter email: ")
  emails.append(e)

# step 1:count domains
d_count={}

for e in emails:
  domain=e.split("@")[1]
  d_count[domain]=d_count.get(domain,0)+1

# step 2: calculate percentage
total=len(emails)

for domain,count in d_count.items():
  percentage=(count/total)*100
  print(f"{domain}: {percentage:.0f}%")
```

5. Problem Statement :

    Sales Spike Detection

```python
# list to store sales
sales = []

n = int(input("Enter number of days: "))

# taking sales input
for i in range(n):
    value = int(input(f"Enter sales for day {i+1}: "))
    sales.append(value)  #this values will be added to sales list

# Step 1: calculate average sales
avg_sales = sum(sales) / len(sales)  #total sales/no. of days

# Step 2: detect spikes (30% above average)
limit = avg_sales * 1.30

# Step 3: display spike days
for i, value in enumerate(sales):
    if value > limit:
        print(f"Day {i+1}: {value}")
```

6. Problem Statement:

   Duplicate User ID Detection

```python
# taking user id
id=input("Enter your user id: ").split()

# occurences
freq={}
for uid in id:
  freq[uid]=freq.get(uid,0)+1

# finding duplicates
duplicates = {uid: freq for uid, freq in freq.items() if freq > 1}

# display result
if duplicates:
    print("Duplicate User IDs:")
    for uid, freq in duplicates.items():
        print(uid, "appears", freq, "times")
else:
    print("No duplicate IDs found")
```
